```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Suite of Food Security Indicators

```python
df_fs = pd.read_csv('https://raw.githubusercontent.com/ahmadhlawa/palestine-food-security-analysis/refs/heads/main/data/Suite%20of%20Food%20Security%20Indicators.csv')

display(df_fs.head())

df_fs.info()
```

```python
df_fs['Value'] = df_fs['Value'].astype(str).str.replace('<', '').str.replace('>', '').str.replace(',', '')
df_fs['Value'] = pd.to_numeric(df_fs['Value'], errors='coerce')
df_fs['Year'] = df_fs['Year'].astype(str)

df_fs_clean = df_fs.dropna(subset=['Value'])


print(df_fs_clean[['Year', 'Value']].isnull().sum())

display(df_fs_clean[['Item', 'Year', 'Value']].head())
```

```python

df_fs_clean['Year'] = df_fs_clean['Year'].astype(str).str.split('-').str[-1].astype(int)


df_fs_pivot = df_fs_clean.pivot_table(index='Year', columns='Item', values='Value').reset_index()

print("شكل الجدول الجديد (الصفوف، الأعمدة):", df_fs_pivot.shape)
print("\nأول 5 صفوف بعد تحويل المؤشرات إلى أعمدة:")
display(df_fs_pivot.head())

print("\nقائمة المؤشرات التي أصبحت أعمدة:")
print(df_fs_pivot.columns.tolist())
```

```python
df_fs.to_csv('Suite of Food Security Indicators_cleaned.csv', index=False)
```

# Data Cleaning & Structural Transformation: Food Security Indicators
In this stage, we processed the FAO Food Security dataset to make it compatible with machine learning models.

Value Normalization: Stripped non-numeric symbols (e.g., '<', '>', ',') and handled suppressed data by converting to numeric types and removing nulls.

Temporal Alignment: Cleaned the 'Year' column by resolving interval formats (e.g., 2018-2020) to single representative integers to allow merging with other datasets.

Feature Engineering (Pivoting): Transformed the data from a long format to a wide format, turning individual food security metrics into independent features (columns) for time-series analysis.
